```python
#public, private, protected access specifiers
#class methods can be used when when instance variable is not required.
#static class

class Employee:
    no_of_leaves=8 #public variable
    
    _proc=9 #protected variable
    __pri=10 #private variable
    #constructor
    def __init__(self,aname,asalary,arole): 
        self.name=aname
        self.salary=asalary
        self.role=arole
   
    
    @classmethod
    def change_leaves(cls, newleaves):
        cls.no_of_leaves=newleaves
    
    @classmethod
    def from_dash(cls,string):
        return cls(*string.split("-"))
    
    @staticmethod
    def printgood(string):
        print("This is good " + string)
        return "Return value of Marry"
        
dipti=Employee("Dipti", 50000,"SD") 
print(dipti._proc)
#print(dipti.__pri)  #gives error
print(dipti._Employee__pri) #Name angling:Because Employee has private member and that class can only access its private variable
```

```python
class A:
    classvar1="I am a class variable in class A"
    def __init__(self):
        self.var1="I am inside class A's constructor"
        self.classvar1="Instance var in class A" #overriding class variable of class B
        self.special="special"
class B(A):
     classvar1="I am in class B" 
     def __init__(self):  #Overriding method : Overrides method of class A
        #super().__init__() #super class constructor can be used in class B
        self.var1="I am inside class B's constructor"
        self.classvar1="Instance var in class B" 
        super().__init__()
        print(super().classvar1)
    
a=A()
b=B()

print(b.special)
print(b.classvar1) 
print(b.var1)
```
