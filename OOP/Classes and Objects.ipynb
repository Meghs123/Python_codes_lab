```python
#self and __init__
class Employee:
    no_of_leaves=8 
    def printdetails(self):
        return f"The Name is {self.name} .Salar is {self.salary} and role is {self.role}"
dipti=Employee()
anjali=Employee()

dipti.name="Dipti"
dipti.salary=50000
dipti.role="Software Developer"

anjali.name="Anjali"
anjali.salary=60000
anjali.role="ML Engineer"

print(dipti.printdetails())
print(anjali.printdetails()) #Benefit of using self is code reusability.
```

```python
#class method
class Employee:
    no_of_leaves=8 
    #constructor
    def __init__(self,aname,asalary,arole): 
        self.name=aname
        self.salary=asalary
        self.role=arole
    #class method
        
dipti=Employee("Dipti", 50000,"SD") 
anjali=Employee("Anjali",60000,"ML")

Employee.no_of_leaves=89
print(dipti.no_of_leaves) #First check for instance variable ,if not found then checks for class variable.

```

```python
#class method
class Employee:
    no_of_leaves=8 
    #constructor
    def __init__(self,aname,asalary,arole): 
        self.name=aname
        self.salary=asalary
        self.role=arole
    #class method decorator
    
    @classmethod #can be used as alt constructor.
    def change_leaves(cls, newleaves): #cls is used for dipti
        cls.no_of_leaves=newleaves
        
dipti=Employee("Dipti", 50000,"SD") 
anjali=Employee("Anjali",60000,"ML")

dipti.change_leaves(56) # we can also use Employee.change_leaves(67)
print(dipti.no_of_leaves) #First check for instance variable ,if not found then checks for class variable.

```

```python
#class methods can be used when when instance variable is not required.
#static class

class Employee:
    no_of_leaves=8 
    #constructor
    def __init__(self,aname,asalary,arole): 
        self.name=aname
        self.salary=asalary
        self.role=arole
   
    
    @classmethod
    def change_leaves(cls, newleaves):
        cls.no_of_leaves=newleaves
    
    @classmethod
    def from_dash(cls,string):
        return cls(*string.split("-"))
    
    @staticmethod
    def printgood(string):
        print("This is good " + string)
        return "Return value of Marry"
        
dipti=Employee("Dipti", 50000,"SD") 
anjali=Employee("Anjali",60000,"ML")
karan=Employee.from_dash("Karan-70000-student")

print(karan.printgood("Marry"))
                         

```
