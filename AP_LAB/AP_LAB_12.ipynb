Q.1) Write a python program to create a class Product having PID, P_Name, price. Now initialize the details of the product and input the quantity of the product, then calculate the total price and display all the details of product.(define calc() and display())

```python
class Product:
  def __init__(self, id, name, price, quantity):
    self.pid = id
    self.pname = name
    self.pprice = price
    self.pquantity = quantity

  def display(self):
    print("Product ID: " + str(self.pid))
    print("Product Name: " + self.pname)
    print("Product Price: " + str(self.pprice))
    print("Product Quantity: " + str(self.pquantity))

  def calc(self):
    self.totalPrice = (self.pprice * self.pquantity)
    print("Total Price: " + str(self.totalPrice))

book_quantity = int(input("Enter quantity of book: "))
watch_quantity = int(input("Enter quantity of watch: "))
toys_quantity = int(input("Enter quantity of toys: "))

book= Product(112233, "book", 100, book_quantity)
watch= Product(1111222, "watch",900, watch_quantity)
toys= Product(111111, "toys", 500, toys_quantity)

book.display()
print()
watch.display()
print()
toys.display()
print()

book.calc()
watch.calc()
toys.calc()
```

Q.2) Write a python program to create a class Bank_Account having data members as AC_No, Name, Balance and member functions as disp_info( ), deposit( ), withdraw( ). Initialize account details using constructor and perform all the mentioned operations. In case of withdraw, the minimum balance should be 500, otherwise it displays the message as “Insufficient Balance”.

```python
class Bank:
    def __init__(self, ac_no, name, bal):
        self.ac_no = str(ac_no)
        self.nm = name
        self.bal = bal
    def dispInfo(self):
        print("Account No.: " + self.ac_no)
        print("Name: " + self.nm)
        print("Current Balance: " + str(self.bal))
    def deposite(self, amount):
        if amount <= 0:
            print("Amount should be more than 0")
            return -1
        else:
            self.bal += amount
            print("Amount added to the account")
            print("New amount: " + str(self.bal))
            return 1
    def withdraw(self, amount):
        if amount < 500:
            print("Amount should be more than 500")
            return -1
        else:
            if self.bal < amount:
                print("Insufficient funds: " + str(self.bal))
                return -1
            else:
                self.bal -= amount
                print("Amount debited from the account")
                print("New amount: " + str(self.bal))
                return 1

b1 = Bank("308", "Ram", 5000)
b1.dispInfo()
b1.deposite(100)
b1.withdraw(4000)
b1.dispInfo()
b1.withdraw(1200)
```

Write a python program that creates objects of both Vehicle and Bus classes and calculates the total fare for both.

```python
class Vehicle:
  
  def __init__(self,name,mileage,capacity):
    self.name=name
    self.mileage=mileage
    self.capacity=capacity
  
  def display(self):
    print("Name of vehicle:",self.name)
    print("Mileage of vehicle:",self.mileage)
    print("Capacity of vehicle:",self.capacity)

  def fare(self):
    return self.capacity*100

class Bus(Vehicle):
  def fare(self):
    amount=super().fare()
    amount+=amount*10/100
    return amount
  
volvo=Bus("volvo",15.61,50)
print("Total bus fare is:",volvo.fare())
```
