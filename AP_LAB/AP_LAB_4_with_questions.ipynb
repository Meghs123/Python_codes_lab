Indexing and Slicing

```python
name="Promit"
print(name)
print(name[4])
print(name[-2])
```

```python
#slicing
name="Promit"
sl1=name[0:3]
print(sl1)
```

```python
name="Promit"
sl1=name[:3]
print(sl1)
```

```python
name="Promit"
sl1=name[1:]
print(sl1)
```

```python
name="Promit"
sl1=name[-3:-1]
print(sl1)    #end term not included
```
