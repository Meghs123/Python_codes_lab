Q.3) Write a python program to input 3 numbers and find out the smallest one using nested if statement.

```python
p=int(input("Enter the 1st number: "))
q=int(input("Enter the 2nd number: "))
r=int(input("Enter the 3rd number: "))
if((p<q) and (p<r)):
  print("Smallest is: ",p)
elif((q<r) and (q<p)):
  print("Smalles is: ",q)
else:
  print("Smallst is: ",r)
```

Q.5) Write a python program to input a 3-digit number and check whether it is palindrome number or not. (without using loop)

```python
s=input(("Enter 3 digit number: "))
if(s==s[::-1]):
  print("The number  is palindrome")
else:
  print("Not a palindrome")
```

Q.6) Write a python program to input three numbers as a,b,c, then display sum of all sides if those are sides of a triangle else display 0 (Zero) on the screen.

Q.7) Write a python program to input marks in 4 subjects, then calculate total mark and percentage. Also display the grade according to the following criteria:
Marks Grade
> 90 A
> 80 and <= 90 B
>= 60 and <= 80 C
below 60 D

```python
m1=int(input("Enter marks 1 out of 50:"))
m2=int(input("Enter marks 2 out of 50:"))
m3=int(input("Enter marks 3 out of 50:"))
m4=int(input("Enter marks 4 out of 50:"))
Tot_marks=m1+m2+m3+m4
Perc=(Tot_marks/4)*2
print("Total Marks:",Tot_marks)
print("Percentage:",Perc)
if(Perc>90):
  print("Grade:A")
elif(Perc>80 and Perc<=90):
  print("Grade:B")
elif(Perc>60 and Perc<=80):
  print("Grade:C")
else:
  Print("Grade:D")
```
