 1) Write a python program to find the area and circumference of a circle according to the given radius value. 

```python
a=int(input("Enter Radius of the circle: "))
area=3.14*a*a;
circumference=2*3.14*a;
print("Area of the circle: ",area)
print("Circumference of the circle: ",circumference)
```

 2) Write a python program to calculate the simple interest of the given P,T,R.

```python
a=int(input("Enter the principal amount: "))
b=int(input("Enter thetime period: "))
c=int(input("Enter the rate of interest: "))
simple_interest=(a*b*c)/100
print("Simple interest of the given P,T,R : ",simple_interest)
```

4) Write a python program to input a value as seconds and convert it into hour, minute and seconds (using operators). 

```python
d=int(input("Enter seconds: "))
min=d//60
sec=d%60
hr=min//60
mins=min%60
print("hour: ",hr,"minutes: ",mins,"sec: ",sec)
```

5) Write a python program to input a 3-digits number, then find and display its reverse (using operators).

```python

x=int(input("Enter a 3-digit no. : "))
d=x%10
num=d*10
x=x//10
d=x%10
num=(num+d)*10
x=x//10
d=x%10
num=num+d
print("Reverse of the number is : ", num)
```

7) Write a python program to input two numbers and find the biggest one (using operators).

```python
a=int(input("Enter 1st number : "))
b=int(input("Enter 2nd number : "))
m=(a*(a//b))+(b*(b//a))
print("Greatest of the two is = ", m)
```
