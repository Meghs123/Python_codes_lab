```python
#factorial
def fibo(n):
        
        if n==0:
            return 1
        else:
            return n*fibo(n-1)
n=int(input("Enter: "))
print(fibo(n))
```

```python
#fibonacci
n=0
p=1
fibo=1
while (fibo<10):
    x=n+p
    print(x)
    fibo=x+fibo
    n,p=p,x
```

```python
#multiply by 2
n=0
p=1
fibo=1
while (fibo<90):
    x=n+p
    print(x)
    fibo=x+fibo
    x,p=p,fibo
```

```python
#sum of the number using recursion
def sum(n):
        
        if n==0:
            return 0
        small_num=sum(n-1)
        output=small_num+n
        return output
n=int(input("Enter: "))
print(sum(n))
```

```python
#raise to the power
x=9
n=2
ans=1
while (n>0):
    ans=ans*x
    n=n-1
print(ans)
```
